```python

def get_model(src_lang,trg_lang):
    src = src_lang
    trg = trg_lang
    
    mname = f'Helsinki-NLP/opus-mt-{src}-{trg}'
    model = MarianMTModel.from_pretrained(mname)
    tok = MarianTokenizer.from_pretrained(mname)
    return model,tok

def translate(model,tok,src_text):
    batch = tok.prepare_translation_batch(src_texts = [src_text])
    gen = model.generate(**batch)
    words: List[str]= tok.batch_decode(gen,  skip_special_tokens=True)
    return words

```

```python
model,tok = get_model('en','fr')
```

```python
word1 = translate(model,tok,["Do you want to play with me?", "say hey","what are you"])
word1
```

## Dynamically get supported languages (English to other language support)

```python

from transformers.hf_api import HfApi
model_list = HfApi().model_list()
org = "Helsinki-NLP/opus-mt-en"
model_ids = [x.modelId for x in model_list if x.modelId.startswith(org)]
suffix = [x.split('-')[4] for x in model_ids]
multi_models = [f'{org}/{s}' for s in suffix if s != s.lower()]
suffix
```
